# Switching between Colour Spaces (RGB, BGR, Grayscale, HSV and Lab) 

```python
import cv2 as cv
img = cv.imread("F:\Image\Cat\cat2.jpg")
cv.imshow('cat', img)
cv.waitKey(0)
```

```python
# BGR to Grayscale
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow('Gray', gray)
cv.waitKey(0)
```

```python
# BGR to HSV
hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
cv.imshow('HSV', hsv)
cv.waitKey(0)
```

```python
# BGR to L*a*b
lab = cv.cvtColor(img, cv.COLOR_BGR2LAB)
cv.imshow('LAB', lab)
cv.waitKey(0)
```

```python
# BGR to RGB
rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
cv.imshow('RGB', rgb)
cv.waitKey(0)
```

```python
# Note: OpenCV read images in the form of BGR, 
# as other like Matplotlib, read the images in the form of RGB
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)                 # BGR
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.imshow(rgb)                # RGB
plt.show()
```

```python
# HSV to BGR
lab_bgr = cv.cvtColor(lab, cv.COLOR_LAB2BGR)
cv.imshow('LAB --> BGR', lab_bgr)
cv.waitKey(0)
```
